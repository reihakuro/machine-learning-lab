```python
import os
import glob
from ultralytics import YOLO

model_path = "/kaggle/input/models/trungkientrungkien/yolo11s/pytorch/default/1/best.pt"
model = YOLO(model_path)

dataset_images_path = "/kaggle/input/datasets/alincijov/self-driving-cars/images"

sample_images = glob.glob(os.path.join(dataset_images_path, "*.*"))[:10]

results = model.predict(
    source=sample_images,
    conf=0.25,       
    save=True,        
    project="/kaggle/working/runs/detect",
    name="baitap7_predict"
)

print("Đã hoàn thành! Kết quả ảnh được lưu tại: /kaggle/working/runs/detect/baitap7_predict")
```
